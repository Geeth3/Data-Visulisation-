```python
import pandas as pd
import os as os 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
MDf2= pd.read_csv('MDF.csv')
MDf2
```

```python
print(MDf2.isnull().sum())
```

```python
result = MDf2.dtypes
print(result)
```

```python
MDf2[['ETHNICITY']] =  MDf2[['ETHNICITY']].fillna('Not stated')
MDf2[['DOD']] =  MDf2[['DOD']].fillna('Alive')
MDf2[['Passed_Vs_Adm']] =  MDf2[['Passed_Vs_Adm']].fillna('Alive')
MDf2[['MED_HX_NAME']] =  MDf2[['MED_HX_NAME']].fillna(0)
MDf2['HEIGHT']=MDf2['HEIGHT'].fillna(MDf2['HEIGHT'].mean())
MDf2['WEIGHT']=MDf2['WEIGHT'].fillna(MDf2['WEIGHT'].mean())
MDf2['DBP']=MDf2['DBP'].fillna(MDf2['DBP'].mean())
MDf2['HR']=MDf2['HR'].fillna(MDf2['HR'].mean())
MDf2['O2SAT']=MDf2['O2SAT'].fillna(MDf2['O2SAT'].mean())
MDf2['Creatinine']=MDf2['Creatinine'].fillna(MDf2['Creatinine'].mean())
MDf2['Glucose']=MDf2['Glucose'].fillna(MDf2['Glucose'].mean())
MDf2['Haemoglobin']=MDf2['Haemoglobin'].fillna(MDf2['Haemoglobin'].mean())
MDf2['Ionised calcium']=MDf2['Ionised calcium'].fillna(MDf2['Ionised calcium'].mean())
MDf2['Lactate']=MDf2['Lactate'].fillna(MDf2['Lactate'].mean())
MDf2['White cell']=MDf2['White cell'].fillna(MDf2['White cell'].mean())
print(MDf2.isnull().sum())
```

```python
MDf2.to_csv('MDf_NAFilled.csv') 
```
